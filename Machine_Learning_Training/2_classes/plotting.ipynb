```python
import matplotlib.pyplot as plt
import numpy as np

# Data
data_types = ['Z', 'ZS', 'T', 'ZST', 'F', 'ZSF', 'ZSEF']
metrics = {
    'Accuracy': [0.908, 0.908, 0.824, 0.910, 0.796, 0.912, 0.915],
    'Precision': [0.824, 0.799, 0.578, 0.793, 0.285, 0.807, 0.776],
    'Recall': [0.694, 0.727, 0.476, 0.749, 0.004, 0.745, 0.813],
    'F1 Score': [0.753, 0.761, 0.522, 0.770, 0.009, 0.775, 0.794],
    'AUC': [0.947, 0.948, 0.808, 0.951, 0.680, 0.950, 0.954]
}

# Plot settings
bar_width = 0.1  # Narrower bars to fit all data types
index = np.arange(len(metrics))  # One position for each metric
n_data_types = len(data_types)

# Create figure and axis
plt.figure(figsize=(14, 6))

# Plot each data type across metrics
for i, data_type in enumerate(data_types):
    # Extract values for this data type across all metrics
    values = [metrics[metric][i] for metric in metrics.keys()]
    plt.bar(index + i * bar_width, values, bar_width, label=data_type)

# Customize plot
plt.xlabel('Metrics')
plt.ylabel('Score')
plt.title('Performance Metrics Across Data Types')
plt.xticks(index + bar_width * (n_data_types - 1) / 2, metrics.keys())
plt.legend(title='Data Type')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.ylim(0, 1)  # Set y-axis range from 0 to 1
plt.tight_layout()

# Show plot
plt.show()
```
