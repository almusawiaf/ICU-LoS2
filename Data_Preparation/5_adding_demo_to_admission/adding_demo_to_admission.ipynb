```python
import pandas as pd
```

```python
MIMIC_Path = '/lustre/home/almusawiaf/PhD_Projects/MIMIC_resources'
df_admissions = pd.read_csv(f'{MIMIC_Path}/ADMISSIONS.csv')
df_patients = pd.read_csv(f'{MIMIC_Path}/PATIENTS.csv')
```

```python
# Merge the two DataFrames based on SUBJECT_ID
df_admission_demo = pd.merge(df_admissions, df_patients[['SUBJECT_ID', 'GENDER', 'DOB', 'DOD', 'DOD_HOSP', 'DOD_SSN', 'EXPIRE_FLAG']], 
                             on='SUBJECT_ID', how='left')

df_admission_demo.head(3)
```

```python
df_admission_demo.to_csv(f'{MIMIC_Path}/ADMISSIONS_DEMO.csv')
```
