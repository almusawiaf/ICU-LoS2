```python
# (A) Removing Features with Excessive Missingness
# If a feature has >80% missing values, consider removing it, unless it holds clinical significance.

drop_cols = ['Platelets', 'Troponin', 'Tidal Volume', 'PEEP Set', 'PTT']
df.drop(columns=drop_cols, inplace=True)


# ==================================================================================================
# B) Filling Missing Values for Clinical Variables
# Numerical Features (Vitals, Labs)

# Use median imputation for robust handling:

num_cols = ['Heart Rate', 'Systolic Blood Pressure', 'Diastolic Blood Pressure',
            'Respiratory Rate', 'Pulse Oximetry', 'Temperature (F)', 
            'WBC', 'RBC', 'Chloride', 'Creatinine', 'Glucose', 'Sodium', 
            'BUN', 'Anion Gap', 'INR']

for col in num_cols:
    df.loc[:, col] = df[col].fillna(df[col].median())

# ==================================================================================================
# (C) Categorical Variables
# Inspired O₂ Fraction (FiO2)

# Use mode imputation:

df.loc[:, 'Inspired O2 Fraction'] = df['Inspired O2 Fraction'].fillna(df['Inspired O2 Fraction'].mode()[0])

# verifying:
assert df.isnull().sum().sum() == 0, "There are still missing values!"
```
