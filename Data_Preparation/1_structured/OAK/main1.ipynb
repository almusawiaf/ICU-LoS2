```python
import pandas as pd
```

```python
df = pd.read_csv('../MIMIC_resources/CHARTEVENTS.csv')
```

```python
df.head(6)
```

```python
df = df.drop(columns=["ERROR", "WARNING", "RESULTSTATUS", "STOPPED", "VALUEUOM"])
df.head()
```

```python
print(df.info())
df = df.dropna(subset=["SUBJECT_ID", "HADM_ID", "ICUSTAY_ID", "ITEMID"])
df.info()
```

```python
all_features = df['ITEMID'].unique()
len(all_features)
```

```python
# Convert CHARTTIME to datetime for proper sorting
df["CHARTTIME"] = pd.to_datetime(df["CHARTTIME"])

# Sort by CHARTTIME so the latest record appears last
df_sorted = df.sort_values(by="CHARTTIME", ascending=True)

# Drop duplicates, keeping the last (most recent) occurrence per group
df_unique = df_sorted.drop_duplicates(subset=["SUBJECT_ID", "HADM_ID", "ICUSTAY_ID", "ITEMID"], keep="first")

# Display result
print(df_unique.info())
```

```python
df_unique.to_csv('../MIMIC_resources/CHARTEVENTS_first_only.csv', index=False)
```
