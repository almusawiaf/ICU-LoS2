```python
import pandas as pd

# Define model names and file paths
summarizing_models = ['1_t5_small2', '3_bart_large_cnn', '4_medical_summarization']
base_path = "../../../Data/unstructured/summarized"

# Load the first model as the base DataFrame
df_main = pd.read_csv(f'../../../Data/unstructured/text/ALL_first_last.csv')

for mdl in summarizing_models:
    df_summary = pd.read_csv(f'{base_path}/{mdl}.csv')
    print(df_summary.keys())
    
    # Rename the SUMMARY column to match the model name
    df_summary = df_summary.rename(columns={'SUMMARY': f'{mdl}_SUMMARY'})
    print(df_summary.keys())
    
    # Merge with the main DataFrame
    df_main = pd.merge(df_main, df_summary[['SUBJECT_ID', 'HADM_ID', f'{mdl}_SUMMARY']], 
                       on=['SUBJECT_ID', 'HADM_ID'], how='inner')

# Save the final merged DataFrame
df_main.to_csv("../../../Data/unstructured/summarized/merged_summaries.csv", index=False)

# Display shape to confirm merging
df_main.head()
```
